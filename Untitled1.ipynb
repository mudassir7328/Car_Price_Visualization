```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load dataset
df = pd.read_csv("car_prices.csv")
#print(df.dtypes)
#print(df.head(11))
```

```python
df["sellingprice"] = pd.to_numeric(
    df["sellingprice"],
    errors="coerce"
)

df["odometer"] = pd.to_numeric(
    df["odometer"],
    errors="coerce"
)

df["condition"] = pd.to_numeric(
    df["condition"],
    errors="coerce"
)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df["sellingprice"] = df["sellingprice"].fillna(
    df["sellingprice"].mean()
)

df["odometer"] = df["odometer"].fillna(
    df["odometer"].mean()
)

df["condition"] = df["condition"].fillna(
    df["condition"].mean()
)
```

```python
cat_cols = df.select_dtypes(include=["object"]).columns

for col in cat_cols:
    df[col] = df[col].fillna(
        df[col].mode()[0]
    )

```

```python
print(df.dtypes)
print(df.head(11))
```

```python
brand_count = df["make"].value_counts().head(10)

plt.figure(figsize=(10,5))
plt.bar(brand_count.index, brand_count.values)
plt.xticks(rotation=45)
plt.xlabel("Brand")
plt.ylabel("Count")
plt.title("Top 10 Car Brands")
```

```python
condition_price = df.groupby(
    "condition"
)["sellingprice"].mean()
plt.figure(figsize=(10,5))
plt.bar(
    condition_price.index.astype(str),
    condition_price.values
)
plt.xlabel("Condition")
plt.ylabel("Average Selling Price")
plt.title("Condition vs Selling Price")
plt.tight_layout()
```

```python
plt.figure(figsize=(10,5))
plt.hist(
    df["sellingprice"],
    bins=20,
    color="purple",
    edgecolor="black"
)
plt.xlabel("Selling Price")
plt.ylabel("Frequency")
plt.title("Selling Price Distribution")
```

```python
color_count = df["color"].value_counts().head(5)
plt.figure(figsize=(10,5))
plt.bar(
    color_count.index,
    color_count.values
)
plt.xticks(rotation=30)
plt.title("Most Common Car Colors")
plt.show()
```
